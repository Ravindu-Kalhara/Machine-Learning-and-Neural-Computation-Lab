# Decision Tree

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
import warnings
warnings.filterwarnings("ignore")
```

Let's use iris dataset.

```python
iris_dataset = load_iris()

X = iris_dataset.data
y = iris_dataset.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
decision_tree = DecisionTreeClassifier(criterion="entropy")

decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
```

```python
plt.figure(figsize=(12, 7))
plot_tree(decision_tree, filled=True, feature_names=iris_dataset.feature_names, class_names=iris_dataset.target_names)
plt.savefig("./imgs/iris_decision_tree.png", dpi=600)
plt.show()
```

```python
print(decision_tree.score(X_test, y_test))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(cm)
disp.plot()
disp.figure_.set_size_inches(w=5, h=5)
```

Let's use Diagnostic Wisconsin Breast Cancer dataset.

```python
cancer_dataset = load_breast_cancer()

X = cancer_dataset.data
y = cancer_dataset.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
decision_tree = DecisionTreeClassifier(criterion="entropy")

decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
```

```python
plt.figure(figsize=(12, 8))
plot_tree(decision_tree, filled=True, feature_names=cancer_dataset.feature_names, class_names=cancer_dataset.target_names)
plt.savefig("./imgs/breast_cancer_decision_tree.png", dpi=600)
plt.show()
```

```python
print(decision_tree.score(X_test, y_test))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(cm)
disp.plot()
disp.figure_.set_size_inches(w=4, h=4)
```
